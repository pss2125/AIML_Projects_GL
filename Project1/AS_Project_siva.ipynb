```python
# Convert the ‘Funding’ features to a numerical value
from re import search
def calcFunding(x):
  x = str(x)
  x = x.lower()
  x = x.replace("$", "")
  if search("k",x):
    x = x.replace("k", "")
    x = float(x)*1000
  elif search("m",x):
    x = x.replace("m", "")
    x = float(x)*1000000
  elif search("b",x):
    x = x.replace("b", "")
    print('billion')
    x = float(x)*1000000000
  return str(x)

cdata_dropna['Funding'] = cdata_dropna['Funding'].apply(calcFunding)
cols = ['Funding']
cdata_dropna[cols] = cdata_dropna[cols].apply(pd.to_numeric, errors='coerce', axis=1)
print(cdata_dropna.info())
```

> Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down? Write the null hypothesis and alternative hypothesis.Test for significance and conclusion



<h3>First We Will see using original data where we havent removed any outliers</h3>

```python
# two-sample t-test
# null hypothesis: the two groups have the same mean
# this test assumes the two groups have the same variance...
```

<h2> Null Hypothesis H0: the two groups have the same mean </h2>
<h2> Alternative Hypothesis H1: the two groups have different mean</h2>
<h2>Significance Level 95%, Alpha = .05</h2>

```python
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power
```

```python
#company data with removing nulls
cdata_dropna.info()
```

```python
#Getting Funding of Operating and Closed companies
operCompanies = cdata_dropna[cdata_dropna['OperatingState'] == 'Operating']['Funding']
closCompanies = cdata_dropna[cdata_dropna['OperatingState'] == 'Closed']['Funding']
```

```python
print('Operating Companies Mean is :' ,operCompanies.mean(), 'and Variance is :', operCompanies.std())
print('Closed Companies Mean is :' ,closCompanies.mean(), 'and Variance is :', closCompanies.std())
```

```python
#Calculating two tailed two sampled t test for Funds of Operating and Closed Companies
#variance is not same
t_statistic, p_value = ttest_ind(operCompanies, closCompanies, equal_var=False)
print(t_statistic, p_value)
```

Based on above Two Sample t - test, **p value is 0.007898898722704608** which is less than .05 sigificance level. Hence **Null Hypothesis is rejected** and **Alternative Hypothesis accepted**. Both operating and closed have significant difference in the funding raised.

<h2> <b>Null Hypothesis H0: </b></h2><h3>The Proportions of Winners that are Operating and Contestants that are Operating is same</h3>

<h3> Significance Level :<b>95% </b></h3><h3>Alpha : <b>.05</b></h3>








```python
from statsmodels.stats.proportion import proportions_ztest

stat, pval = proportions_ztest([len(winnersOper), len(contestantsOper)] , [len(winners), len(contestants)])

if pval < 0.05:
  print('p value is ', "\033[1m",round(pval,4), "\033[0m", ". The p value is lesser than 0.05. Hence we reject the Null Hypothesis")
  #print(f'With a p-value of {round(pval,4)} the difference is significant. aka |We reject the null|')
else:
  print('p value is ', "\033[1m",round(pval,4), "\033[0m", ". The p value is greater than 0.05. Hence we retain the Null Hypothesis")
  #print(f'With a p-value of {round(pval,4)} the difference is not significant. aka |We fail to reject the null|')
```
